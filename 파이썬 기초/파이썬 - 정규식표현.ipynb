```python
# 01-re1.py

data= """
홍길동 123456-1234567
전우치 456789-1234567
유관순 708768-1234567
"""

result =[]
for line in data.split("\n"):
    word_result = []
    for word in line.split(" "):
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit():
            word = word[:7] + "*******"
        word_result.append(word)
    result.append(" ".join(word_result))
    
print("\n".join(result))

#-------------------------
#정규표현식을 이용하면...
#-------------------------

import re

pat = re.compile("(\d{6})[-]\d{7}")
print(pat.sub("\g<1>-*******", data))  #\g 그룹을 의미함

```

```python
# 예제 6 - 그룹핑 '

#------------------------------------------------------
#정규표현식 1 - 그룹핑 옵션
#-------------------------------------------------------

p1 = re.compile('(ABC)+')
m1 = p1.match('ABCABCABC Ok?')
print(m1)
print(m1.group())

#----------------------------
#전화번호 형태의 문자열을 찾는 정규표현식

p21 = re.compile(r'\w+\s+\d+[-]\d+')
m21 = p21.search('홍길동 010-1234-1234')
print(m21)
print(m21.group())

p21 = re.compile(r'(\w+)\s+\d+[-]\d+[-]\d+')
m21 = p21.search('홍길동 010-1234-1234')
print(m21)
print(m21.group()) #매치된 전체 문자열
print(m21.group(0)) #매치된 전체 문자열
print(m21.group(1)) #첫 번째 그룹에 해당되는 문자열


p21 = re.compile(r'(\w+)\s+(\d+[-]\d+[-]\d+)')
m21 = p21.search('홍길동 010-1234-1234')
print(m21)
print(m21.group(2)) #두 번째 그룹에 해당되는 문자열 

#그룹이 중첩되어 사용될 경우에는 바같쪽부터 시작하여 안쪽으로 
#들어갈수록 인덱스가 증가한다. 

p21 = re.compile(r'(\w+)\s+((\d+)[-]\d+[-]\d+)')
m21 = p21.search('홍길동 010-1234-1234')
print(m21)
print(m21.group(2))
print(m21.group(3))

#그룹핑된 문자열 재참조
p3 = re.compile(r'(\b\w+)\s +\1')
m3 = p3.search('Paris in the the spring')
print(m3)
print(m3.group())

#---------------------------------------------------
#그룹핑된 문자열에 이름 붙이기 

p21 = re.compile(r'(\w+)\s+((?P<Guk>\d+)[-]\d+[-]\d+)')
m21 = p21.search('홍길동 010-1234-1234')
print(m21)
print(m21.group('Guk'))
print(m21.group(3))

#정규식 내에서 그룹명을 이용한 재참조
p3 = re.compile(r'(?P<w1>\b\w+)\s+(?P=w1)')
m3 = p3.search('Paris in the the spring')
print(m3)
print(m3.group())
```

```python
#예제 7 - 전방탐색

#------------------------------------------------------
#정규표현식 1 - 전방 탐색 
#-------------------------------------------------------

p1 = re.compile('.+:')
m1 = p1.match('http://google.com')
print(m1)
print(m1.group())

#--------------------------------------------------------
#긍정형 전방 탐색 
#해당하는 정규식과 매치되어야 하며 조건이 통과되어도 문자열이 소모되지 않음
#**검색 조건에는 추가되지만 검색 결과에는 추가되지 않음.

p2 = re.compile('.+(?=:)')
m2 = p2.match('http://google.com')
print(m2)
print(m2.group())

#--------------------------------------------------------

data = """
foo.bar
autoexec.bat
autoexec.exe
sendmail.cf
"""

#파일명 + '.' + 확장자를 나타내는 정규식 
p31 = re.compile('.*[.].*$', re.MULTILINE)
m31 = p31.findall(data)
print(m31)

#확장자가 bat 인 파일은 제외하기
#확장자가 b로 시작하는 것 제외
# - foo.bar도 제외.
p32 = re.compile('.*[.][^b].*$', re.MULTILINE)
m31 = p32.findall(data)
print(m31)

#부정형 전방 탐색
#확장자가 bat 인파일은 제외하기
p41 = re.compile('.*[.](?!bat$).*$', re.MULTILINE)
m41 = p42.findall(data)
print(m41)

p42 = re.compile('.*[.](?!bat$|exe$)..*$', re.MULTILINE)
m42= p42.findall(data)
print(m42)
```

```python
# 예제 8 - 문자열 바꾸기

#------------------------------------------------------
#정규표현식 1 - 문자열 바꾸기 
#-------------------------------------------------------

p1 = re.compile('(blue|white|red)')
m11 = p1.sub('color', 'blue socks and red shoes')
print(m11)

#횟수 제함. 딱 1번만 바꾸기
m12 = p1.sub('color', 'blue socks and red shoes', count =1)
print(m12)

#리턴을 튜플로 반환하는 subn
m13 = p1.subn('color', 'blue socks and red shoes')
print(m13)

#sub 메서드 사용시 참조 구문 사용하기
p2 = re.compile(r'(?P<name>\w+)\s+(?P<phone>\d+[-]\d+[-]\d+)')

#그룹 이름으로
m21 = p2.sub('\g<phone> \g<name>', '홍길동 010-1234-4567')
print(m21)

#그룹 인덱스로
m22= p2.sub('\g<2> \g<1>', '홍길동 010-1234-4567')
print(m22)

#sub 메서드의 입력 인수로 함수 넣기
#Return the hex string for a decimal number
def hexrepl(match):
    value = int(match.group())
    return hex(value)

p3 = re.compile(r'\d+')
m3 = p3.sub(hexrepl, 'Call 65470 for printing, 764515 for user code.')
print(m3)

#---------------------------------------------
#Non- Greedy vs Greedy

s1 = "<html><head><title>Python</title></head>"
print(len(s1))

#메타문자는 매우 탐욕스러워 매치할 수 있는 최대한의 문자열을 모두 소모
s2 = re.match('<.*>', s1)
print(s2.span())
print(s2.group())

#Non- Greedy 문자인 ? 을 사용 
#가능한한 최소한의 반복을 수행하도록 도와주는 역할을 한다. 
s2 = re.match('<.*?>', s1)
print(s2.span())
print(s2.group())
```

```python
#28_NaverRestAPI.py

import urllib.request
import pandas as pd 
import json
import re

def clean_html(x):
    x = re.sub("\&\*\;","",x)
    x= re.sub("<.*?>","",x)
    return x

client_id = "******************Sm"
client_secret = "*************li"

search_word='코로나'  #검색어
encode_type='json'    #출력방식 json 또는 xml
max_display=100       #출력 뉴스 수
sort='date'           #결과값의 정렬 기준 시간순 date,관련도 순 sim
start=1               #출력위치

url="https://openapi.naver.com/v1/search/news.json"
#option="&display=3&sort=count"
option=f"&display={str(int(max_display))}&start={str(int(start))}&sort={sort}"
query="?query="+urllib.parse.quote(search_word)
url_query=url+query+option
print(query)

#Open API검색 요청 개체 설정
request=urllib.request.Request(url_query)
#헤더에 아이디와 키 정보 넣기
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)

#검색 요청 및 처리
response=urllib.request.urlopen(request)
rescode=response.getcode()
#HTTP요청에 대한 정상응답일 경우
if (rescode==200):
    #response.read()메서드로 수신된 HTML데이터를 가져올 수 있음
    response_body=response.read().decode('utf-8')
    #json.loads()메서드를 사용해서 data에 수신된 데이터를
    #사전 데이터로 분석해서 자동으로 만들어줌
    data=json.loads(response_body)
else:
    print("Error Code:"+rescode)
    
#데이터 프레임 만들기
df=pd.DataFrame(data['items'])
df.info()

#특수문자 제거
df['title']=df['title'].apply(lambda x: clean_html(x))
df['description']=df['description'].apply(lambda x:clean_html(x))
print(df.head(10))


print(df)
```

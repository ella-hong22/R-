```python
#01_basic.py

import pandas as pd


#그래프에서 사용할 데이터
input_file= "E:/Ella/RStudio/DataAnalysis/04.matplotlib/Data/04_exam4.csv"
df= pd.read_csv(input_file)

#히스토그램(빈도표)
ax1 = df.math.plot.hist()

ax2 = df.loc[:,["math", "english"]].plot.hist(alpha= 0.5, bins=10) #두개 그래프가 동시에 

#밀집도(선그래프)
ax3 = df["math"].plot.kde()

#산점도
ax4 = df.plot.scatter(x="id",y="math")

#육각 그래프
#산점도 그래프는 점이 겹쳐 보일 경우 점을 구분하기 어렵다는 단점이 있다.
ax5 = df.plot.hexbin(x = "id",y="math", gridsize =10)

#박스
df2 = df.loc[:,["math","english","science"]]
ax6 = df2.plot.box()

#파이
df3 = df["class"].value_counts()
ax7 =df3.plot.pie(autopct="%.2f%%")
```

```python
# 02_anscombe.py

import matplotlib.pyplot as plt
import seaborn as sns

#그래프에서 사용할 데이터 
anscombe = sns.load_dataset("anscombe")
dataset1 = anscombe.loc[anscombe['dataset']== "I"]
dataset2 = anscombe.loc[anscombe['dataset']== "II"]
dataset3 = anscombe.loc[anscombe['dataset']== "III"]
dataset4 = anscombe.loc[anscombe['dataset']== "IV"]

#다음을 통해 평균과 표준편차가 같음
dataset1.describe()
dataset2.describe()
dataset3.describe()
dataset4.describe()
#그래프 격자가 위치할 기본 틀 그림 생성
fig = plt.figure()

#기본 특에 몇 개의 그래프를 그릴지 결정한다. 
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

#그래프의 형태를 지정하고 데이터 세팅
ax1.plot(dataset1['x'], dataset1['y'], 'o')
ax2.plot(dataset2['x'], dataset2['y'], 'o')
ax3.plot(dataset3['x'], dataset3['y'], 'o')
ax4.plot(dataset4['x'], dataset4['y'], 'o')

#개별그래프에 타이틀 지정
ax1.set_title("dataset1")
ax2.set_title("dataset2")
ax3.set_title("dataset3")
ax4.set_title("dataset4") 

#전체 그림 세팅
fig.suptitle("Anscombe Data")
fig.tight_layout()

#그림 저장 
plt.savefig('E:/Ella/RStudio/DataAnalysis/04.matplotlib/Output/01_plot.png', dpi=300, bbox_inches='tight')

#그래프실제로 보여주기
plt.show()
```

```python
#bar 

#기본 속성 지정 
plt.style.use('ggplot')
plt.rc('font', family ='D2Coding')

#그래프에서 사용하러 데이터
customers = ['홍길동', '전우치', '손오공', '해리포터','멀린']
customers_index  = [ 0,1,2,3,4]
sale_amounts = [127,90,202,111,232]

#그래프 격자가 위치할 기본 틀 그림 생성
fig = plt.figure()

#그래프 격자 그리기
ax1 = fig.add_subplot(1,1,1)

#그래프의 형태를 지정하고 덷이터 세팅
ax1.bar(customers_index, sale_amounts, align ='center', color = 'darkblue')

#눈금 위치 표시
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

#막대의 눈금 레이블을 인덱스에서 실제 이름으로 변경
#rotation =0 은 눈금 레이블을 수평으로 지정
plt.xticks(customers_index, customers, rotation=0, fontsize= 'small')

#그래프에서 레이블 지정 
plt.xlabel('고객이름')
plt.ylabel('판매량')
plt.title('Sale Amount per Customer')

plt.show() 
#그래프 저장
plt.savefig('E:/Ella/RStudio/DataAnalysis/04.matplotlib/Output/02_plot.png', dpi=300, bbox_inches = 'tight')

```

```python
#07_line

#기본 속성 지정 
plt.style.use('ggplot')
plt.rc('font', family ='D2Coding')

#그래프에서 사용하러 데이터#그래프에서 사용할 데이터
input_file= "E:/Ella/RStudio/DataAnalysis/04.matplotlib/Data/04_exam4.csv"
df= pd.read_csv(input_file)

dataset1 = df.loc[:,"math"]
dataset2 = df.loc[:,"english"]
dataset3= df.loc[:,"science"]

#그래프 격자가 위치할 기본 틀 그림 생성
fig = plt.figure()

#그래프 격자 그리기
#ax1이 유일한 (1행, 1열, 1개) 하위 그래프임을 의미
ax1 = fig.add_subplot(1,1,1)

#다른 모양의 선 그래프 3ro todtjd
ax1.plot(df.index, dataset1, marker='o', color='blue', linestyle="-", label="Math")
ax1.plot(df.index, dataset2, marker='+', color='red', linestyle="-", label="english")
ax1.plot(df.index, dataset3, marker='*', color='green', linestyle="-", label="science")

major_ticks = range(0,21,5)
minor_ticks = range(0,21,1)

ax1.set_xticks(major_ticks)
ax1.set_xticks(minor_ticks, minor =True)

ax1.grid(which = 'both')

ax1.grid(which = 'minor', alpha=0.2)
ax1.grid(which = 'major', alpha=0.5)


plt.xlabel('학생번호', fontsize = 12)
plt.ylabel('점수', fontsize= 12)
ax1.set_title('Line plots :')

#그래프의 범례를 만든다.
#loc='best'는 그래프에서 최상의 위치에 자동적으로범례가 배치됨.
plt.legend(loc='best', fontsize=10)
plt.savefig('E:/Ella/RStudio/DataAnalysis/04.matplotlib/Output/06_line.png', dpi=300, bbox_inches = 'tight')

plt.show()
```

```python
# 08_scatter

#기본 속성 지정 
plt.style.use('ggplot')
plt.rc('font', family ='D2Coding')

#그래프에서 사용하러 데이터#그래프에서 사용할 데이터
input_file= "E:/Ella/RStudio/DataAnalysis/04.matplotlib/Data/04_exam4.csv"
df= pd.read_csv(input_file)

dataset1 = df.loc[:,["id","math"]]

#그래프 격자가 위치할 기본 틀 그림 생성
fig = plt.figure()

#그래프 격자 그리기
#ax1이 유일한 (1행, 1열, 1개) 하위 그래프임을 의미
ax1 = fig.add_subplot(1,1,1)

#산점도를 만든다. 
ax1.scatter(df.id, df.math, color='red')

#레이블 작성 
plt.xlabel('학생번호', fontsize = 12)
plt.ylabel('점수', fontsize= 12)
ax1.set_title('Line plots :')

#실제 데이터 값을 기반으로 축의 범위를 정하기
plt.xlim((dataset1["id"].min()-1., dataset1["id"].max()+1.))
plt.ylim((dataset1["math"].min()-10., dataset1["math"].max()+10.))

plt.savefig('E:/Ella/RStudio/DataAnalysis/04.matplotlib/Output/07_scatter.png', dpi=300, bbox_inches = 'tight')

plt.show()
```

```python
# 09_scatter2.py

#기본 속성 지정 
plt.style.use('ggplot')
plt.rc('font', family ='D2Coding')

#그래프에서 사용하러 데이터#그래프에서 사용할 데이터
input_file= "E:/Ella/RStudio/DataAnalysis/04.matplotlib/Data/04_exam4.csv"
df= pd.read_csv(input_file)

groups = df.groupby('class')

#그래프 격자가 위치할 기본 틀 그림 생성
fig = plt.figure(figsize=(10,7))

#그래프 격자 그리기
#ax1이 유일한 (1행, 1열, 1개) 하위 그래프임을 의미
ax1 = fig.add_subplot(1,1,1)

#산점도를 만든다. 
for name, group in groups:
    ax1.plot(group.id,
            group.math,
            marker='o',
            linestyle='',
            label=name)
    
#레이블 작성 
plt.xlabel('학생번호', fontsize = 12)
plt.ylabel('점수', fontsize= 12)
ax1.set_title('Scatter plots with best fit lines')


plt.savefig('E:/Ella/RStudio/DataAnalysis/04.matplotlib/Output/08_scatter.png', dpi=300, bbox_inches = 'tight')

plt.show()
```

```python
#10_boxplot

#기본 속성 지정 
plt.style.use('ggplot')
plt.rc('font', family ='D2Coding')

#그래프에서 사용하러 데이터#그래프에서 사용할 데이터
input_file= "E:/Ella/RStudio/DataAnalysis/04.matplotlib/Data/04_exam4.csv"
df= pd.read_csv(input_file)

groups = df.groupby('class')

#그래프 격자가 위치할 기본 틀 그림 생성
fig = plt.figure(figsize=(10,7))

#그래프 격자 그리기
#ax1이 유일한 (1행, 1열, 1개) 하위 그래프임을 의미
ax1 = fig.add_subplot(1,1,1)

#각 상자그림의 레이블
box_labels = ['normal','normal_sample','lognormal','lognormal_sample']

#상자그림을 그린다.
ax1.boxplot([df.loc[:,"math"],
           df.loc[:,"english"],
            df.loc[:,"science"]],
           labels = ["math","english","science"],
           notch = False, sym ='.', vert =True,
           showmeans= True)

#레이블 작성 
plt.xlabel('과목', fontsize = 12)
plt.ylabel('점수', fontsize= 12)
ax1.set_title('box  plots with best fit lines')


plt.savefig('E:/Ella/RStudio/DataAnalysis/04.matplotlib/Output/09_box.png', dpi=300, bbox_inches = 'tight')

plt.show()
#notch = False는 상자 중간에 홈이 아피게 하지 말고 사각형으로 그리게 한다.
#sym ='.'는 특이점을 표현할 기호를 기본값인  + 대신 으로 지정했다.
# vert= True 는 상자를 세로 방향
# showmeans = True는 평균도 표시 여부
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 



#==============================================
#seaborn : histogram =  1차원 실수 분포 데이터
#==============================================

#seaborn은 matplotlib.pyplot을 임포트 해야 쓸 수 있다.
#seaborn은 matplotlib보다 화려한 그래프를 만들 수 있다.

import matplotlib.pyplot as plt
import seaborn as sns

#그래프에서 사용할 데이터
#어떤 식당에서 팁을 지불한 손님의 정보를 모아둔 데이터
#지불금액, 팁, 성별, 흡연유무, 요일, 시간, 전체인원
tips = sns.load_dataset("tips")
print(tips)

#아래 셋중에 하나 선택
#(1) 히스토그램 + 밀집도 그리기
ax = sns.distplot(tips["total_bill"])
#(2) 밀집도(kernel density) 그래프 제외
#ax = sns.distplot(tips["total_bill"], kde=False)
#(3) 히스토그램 그래프 제외
#ax = sns.distplot(tips["total_bill"], hist=False)

#(4) 데이터 시각화의 효율화를 위해 여러 그래프를 한번에 출력할 경우
# = 양탄자(rug) 그래프 추가하여 출력
# = 양탄자 그래프는 그래프의 축에 동일한 길이의 직선을 붙여
# 데이터 밀집정도를 표현한 그래프 
ax = sns.distplot(tips["total_bill"], rug=True)

#레이블 작성
ax.set_title('Total Bill Histogram with Density Plot')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')

#파일로 저장
# plt.savefig('../Output/11_Sns.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 13_sns_scatter


#==============================================
#seaborn : 산점도 + 회귀선 한꺼번에 그리기
#==============================================


#그래프에서 사용할 데이터
tips = sns.load_dataset("tips")

#산점도+ 회귀선 그래프 그리기
ax = sns.regplot(x='total_bill', y='tip', data =tips)

#산점도 그래프만 그리기
# ax = sns.regplot(x = 'total_bill', y ='tip', data =tips, fit_reg= False)

#레이블 작성
ax.set_title('Scatterplot of total bill and tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('tip')
```

```python

#==============================================
#seaborn : 산점도 + 히스토그램 한꺼번에 그리기
#==============================================


#그래프에서 사용할 데이터
tips = sns.load_dataset("tips")

#산점도+ 회귀선 그래프 그리기
ax = sns.jointplot(x='total_bill', y='tip', data =tips)

#레이블 작성 
#다음은 x,y축 레이블은 표시되지만, 전체 타이틀이 표시되지 않는ㄷ.

# ax.set_axis_labels(xlabel='Total Bill', ylabel ='tip')
ax.fig.suptitle('joint plot of total bill and tip', fontsize =10, y = 1.03)
```

```python
# 15_sns_hex.py
#==============================================
#seaborn : 산점도 + 히스토그램 한꺼번에 그리기
#==============================================


#그래프에서 사용할 데이터
tips = sns.load_dataset("tips")

#산점도+ 회귀선 그래프 그리기
ax = sns.jointplot(x='total_bill', y='tip', data =tips, kind='hex')

#레이블 작성 
#다음은 x,y축 레이블은 표시되지만, 전체 타이틀이 표시되지 않는ㄷ.

# ax.set_axis_labels(xlabel='Total Bill', ylabel ='tip')
ax.fig.suptitle('joint plot of total bill and tip', fontsize =10, y = 1.03)
```

```python
# 16_sns_kde.py

ax =sns.kdeplot(data=tips['total_bill'], data2=tips['tip'],  shade =True)

#레이블 작성
ax.set_title('Kernel Density estimation plot of total bill and Tip')
ax.set_xlabel('Totla bill')
```

```python
#17_sns_bar.py

#==============================================
#seaborn : 바 그래프 - 지정한 변수의 평균을 계산하여 그림 
#==============================================

#평균은 막대의 높이로, 편차는 에러바로 표시한다. 
ax = sns.barplot(x='time', y='total_bill', data =tips, hue='sex')

# ax.set_title('Kernel Density estimation plot of total bill and Tip')
plt.title('hi')
ax.set_xlabel('time of day')
ax.set_ylabel('Average total bill')
```

```python
import numpy as np
import seaborn as sns

x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)

plt.plot(x, y)

plt.xlabel('x value', fontsize=15)
plt.ylabel('y value', fontsize=15)
plt.title('sin graph', fontsize=18)

plt.show()
```

```python
#heatmap
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns


#==============================================
#seaborn : 히트맵
#==============================================

#기본 속성 지정 
plt.style.use('ggplot')
plt.rc('font', family ='D2Coding')

#그래프에서 사용하러 데이터#그래프에서 사용할 데이터
input_file= "E:/Ella/RStudio/DataAnalysis/04.matplotlib/Data/04_exam4.csv"
df= pd.read_csv(input_file)
df2 = df.loc[:, ["math", "english","science"]]

#그래프  그리기 (fmt='d' 정수 형태(integer format)로 숫자 입력)
sns.heatmap(df2, annot=True, fmt="d", linewidths =1)

plt.show()

 #피봇테이블을 이용한 히트맵
tips = sns.load_dataset('tips')
tips
pivot = tips.pivot_table(index='day', columns='size', values='tip')
pivot

sns.heatmap(pivot, cmap='Blues', annot=True)
plt.show()
```

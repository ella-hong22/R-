```python
#11-groupby2.py

input_file = "E:/Ella/RStudio/DataAnalysis/01.CSV/Data/csv_exam4.csv"
df = pd.read_csv(input_file)

groupby_class = df.groupby('class')

#특정 그룹 값 가져오기
groupby_class=df.groupby('class')

#특정 그룹 값 가져오기
groupby_class.get_group(1)

#전체 그룹별 값 가져오기 
for name, group in groupby_class :
    print(str(name) + " : " + str(len(group)) )
    print(group)
    print()
groupby_class.head(2).head(10)

#라이브러리에서 제공되는 집계 메서드로 원하는 값으 계산할 수 없는 경우에는 
#직접 만들어서 사용해야한다. 
#agg: 사용자함수와 grouoby 메서드 조합

#집계 방법1 -> {열: 연산} dict를 입력
groupby_class.agg({'math' : 'sum'})

#집계 방법 2 -> 컬럼 선택후 연산자만 전달
groupby_class['math'].sum()

#multi columns and aggs 
groupby_class['math', 'english'].agg('sum')
groupby_class['math', 'english'].agg(['sum', 'mean'])

#사용자 정의 함수를 agg에 활용
def myFunc1(s):
    return s.max()

groupby_class['math'].agg(myFunc1)

#filter : 그룹화한 데이터에서 원하는 데이터를 걸러낼 때 사용 
data = {"class2" : ['A','A','B','B','C','C'],
       "name" : ['apple','banana','orange','pear','melon','peach'],
       "sex" : ['M','F','M',None,'F','F'],
       "math" : [60,70,80,90,100,100]}

df2 = pd.DataFrame(data)
groupby_class2 = df2.groupby('class2')

for name, group in groupby_class2 :
    print(str(name) + " : " + str(len(group)) )
    print(group)
    print()
    
#위의 group, 아래의 x 
filtered = groupby_class2.filter(lambda x : x['math'].max() > 80)
print(filtered)

#다시 집계 메서드를 사용해 주어야 한다. 
filtered['class2'].value_counts()

#apply : 모든 종류의 groupby 객체 처리가 가능. 유연성이 가장 높다. 

#scalar 반화
def myFunc41(df):
    return df['math'].max()

#Series 반환
def myFunc42(df, n) :
    return df[df['class'] == 1]

#groupby 객체에 반영(모두 가능)
groupby_class.apply(myFunc41)
groupby_class.apply(lambda x : myFunc42(x,1))

```

```python
# 14_timeline_exam1.py

#---------------------------------------------------------
# 판다스 : 시계열 데이터 다루기
#---------------------------------------------------------

data = {"date" : ['2020-01-01', '2020-01-03' , '2020-01-04', 
                 '2020-01-05','2020-01-06','2020-01-07'],
       "A" : ['A1','A2','A3','A4','A5','A6'],
       "B" : [None,None, 'B1', 'B2','B3','B4'],
       "C" : [None,None,None,None, 'C1', 'C2']}

df= pd.DataFrame(data)
print(df)
print(df.dtypes)

#데이터 속성 변경 
df['date'] = pd.to_datetime(df["date"])
print(df.dtypes)

#시계열 데이터를 다루기 위해 date 열을 인덱스로 지정
df.index = df['date']
df  = df.drop('date', axis =1)
print(df)

#date 열의 최댓값과 최솟값으로 시간 범위 생성
new_idx = pd.date_range(start= df.index.min(), end= df.index.max())
print(new_idx)

#생성된 인덱스 반대로 
#new_idx = reversed(new_idx)

#새로 생성한 인덱스 적용
df = df.reindex(new_idx)
print(df)

#---------------------------------------------------------------------------------
#각 칼럼의 시작일 옮기기
#last_valid_index 는 가장 오래된 데이터를 가져오고
#first_valid_index 는 가장 최근의 데이터를 가져온다. 

last_valid = df.apply(pd.Series.last_valid_index)
print(last_valid)

first_valid = df.apply(pd.Series.first_valid_index)
print(first_valid)

#각 칼럼의 시작일을 동일한 출발선으로 옮기기
#데이터가 가장 처음 나오는 날에서 각 컬럼의 데이터 시작일을 
#뺀 만큼 옮기면 된다. 

earliest_date = df.index.min()
print(earliest_date)

shift_values = first_valid - earliest_date
print(shift_values)

df['B']
df['B'].shift(3) #컬럼 데이터 아래로 3칸 밀기
df['B'].shift(-3)

df_dict = {}
for idx, col in enumerate(df):
    d = shift_values[idx].days
    print(idx, col, d)
    shifted = df[col].shift(d*-1)
    df_dict[col] = shifted
    
df_shift = pd.DataFrame(df_dict)
print(df_shift)
```

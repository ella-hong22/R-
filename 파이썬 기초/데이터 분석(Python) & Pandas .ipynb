```python
# 01n_simple_parsing_and_write.py

import sys

input_file = "E:/Ella/RStudio/DataAnalysis/01.CSV/Data/01_data.csv"
output_file = "E:/Ella/RStudio/DataAnalysis/01.CSV/Output/out01_data.csv"

with open(input_file, "r", newline ='')as filereader:
    with open(output_file, 'w', newline='')as filewriter:
        #헤더 읽기
        header = filereader.readline()
        header = header.strip()
        header_list = header.split(',')
        print(header_list)
        filewriter.write(','.join(map(str, header_list))+'\n')
        
        #데이터 읽기
        for row in filereader:
            #데이터 쓰기 위한 전처리
            row = row.strip()
            row_list = row.split(',')
            print(row_list)
            filewriter.write(','.join(map(str, row_list))+
                             '\n')
            
            #map() 함수는 header_list의 각 원소에 str() 함수를 적용하여 
            #각 원소를 문자열로 만든다. 
            
```

```python
#02n_parsing_and_write.py
import csv
import sys

#기본 파이썬 csv 모듈 사용 
# 파이썬에 내장된 csv 모듈을 사용하여 csv 파일을 처리할 때의 좋은 점 중하나는 
# 데이터에 포함된 쉼표 및 기타 복잡한 패턴을 정확하게 처리할 수 있다. 

input_file = "E:/Ella/RStudio/DataAnalysis/01.CSV/Data/01_data.csv"
output_file = "E:/Ella/RStudio/DataAnalysis/01.CSV/Output/out02_data.csv"

with open(input_file, "r", newline ='') as csv_in_file:
    with open(output_file, 'w', newline='') as csv_out_file:
        #csv 내장모듈 사용하여 읽고 쓰기 
        filereader = csv.reader(csv_in_file, delimiter = '/')
        filewriter = csv.writer(csv_out_file, delimiter = '/')
        for row_list in filereader:
            print(row_list)
            #데이터가 리스트라면 , 로 붙여진 문자열로 변형하여 
            #파일에 데이터를 wirte 한다.
            #앞의 코드에 비하면 훨씬 간편하다. 
            filewriter.writerow(row_list)
```

```python
#09p1_DataFrame2.py
#DataFrame에 대한 모든 것. 

import pandas as pd

data = {"name" : ["홍길동", "전우치","손오공","해리포터","멀린"],
       "age" : [ 20,25,30,15,40],
       "math" : [90,80,None,80,100],
       "science" : [80,None, 70, None, 100]}

df = pd.DataFrame(data)

#특정 행 또는 특정 열에서만 계산하기
df['age'].sum()
df.loc[:, 'age'].sum()

#행방향으로의 합(즉, 각 열의 합)
df.sum(axis = 0)

#열방향으로의 합 (각 행의 합 )
df.sum(axis =1)

#이때, NaN 값은 배제
#Nan 값을 배제하지 않고 계산하려면 아래와 같이 skipna에 대해 false를 지정해준다. 
df.sum(axis = 1 , skipna = False)


#dataFrame 전체의 결측값 여부 확인
#  : df.isnull(), isnull(df, df.notnull(), notnull(df))

df.isnull().sum()
df['science'].isnull().sum()
df.notnull().sum() 

#'string' 데이터 형식 : 'None'을 할당 --> 'None' 으로 입력
#'float' 데이터 형식 : 'None'을 할당 --> 'NaN' 으로 입력
df.iloc[[0,1],[0,1]] = None
print(df)

#행의 값의 모든 값이 nan인 경우 그 행을 없앰
df.dropna(how='all')

#행의 값중 하나라도 nan인경우 그 행을 없앤다. 
df.dropna(how = 'any')

#nan 값에 값넣기
df.fillna(value = 0.0)

#특정 행 drop 하기
df.drop(0)

#2개 이상도 가능 
df.drop([1,2])

#특정 열 삭제하기
df.drop('science', axis = 1)
#2개 이상의 열도 가능 
df.drop(['science', 'math'], axis =1)

# pandas에서 DataFrame에 적용되는 함수들
# sum() 함수 이외에도
# cumsum 맨 첫 번째 성분부터 각 성분까지의 누적합을 계산 (0에서부터 계속 더해짐)
# cumprod 맨 첫번째 성분부터 각 성분 까지의 누적곱을 계산 (1에서 부터 계속 곱해짐 )


#정렬
# df.sort_values(by = ["grp","val"], ascending = [True, False])

#column을 기준으로?
df.sort_index(axis =1) 

#내림차순으로는?
df.sort_index(axis =1, ascending = False) 

df.sort_values(by='math', ascending = False)

#지정한 행 또는 열에서 중복값을 제외한 유니크한 값만 얻기
df['math'].unique()
df['math'].nunique()

#지정한 행 또는 열에서 값에 따른 개수 얻기
df['math'].value_counts()
df['math'].value_counts(normalize =True)

#ㅣ정한 행 또는 열에서 입력한 값이 있는지 확인
df['math'].isin([80,90])

#열의 값이 alpha 나 beta 인 ㅎ모든 행 구하기
df.loc[df['name'].isin(['홍길동', '손오공']),:]
```

```python
#전체 파일 개수 및 각  파일의 행 및 열 개수 계산 
import glob
import os
import sys

input_path = "E:/Ella/RStudio/DataAnalysis/01.CSV/Data"

file_counter = 0

#해당 디렉터리에서 sales 가 포함된 모든 파일 찾아 오기
for input_file in glob.glob(os.path.join(input_path,'*sales*')):
    row_counter =1
    #개별 파일 열기
    with open(input_file, 'r', newline ='') as csv_in_file:
        filereader = csv.reader(csv_in_file)
        #해당 파일에 헤더가 있으므로 일단 한 줄 읽고 : 리스트 반환
        header= next(filereader)
        #나머지는 내용이므로 줄 수 증가시킴
        for row in filereader:
            row_counter +=1
    #포맷 문자열로 출력
    print('{0!s}: \t{1:d} rows \t{2:d} columns'.format(\
                                                      os.path.basename(input_file),row_counter, len(header)))
    
    file_counter += 1
    
print(' 파일 갯수 : {0:d}'.format(file_counter))
```

```python
#판다스를 이용 여러 파일의 데이터 (Row ) 합치기

input_path = "E:/Ella/RStudio/DataAnalysis/01.CSV/Data"
output_file = "E:/Ella/RStudio/DataAnalysis/01.CSV/Output/out11p.csv"

#해당 디렉터리에서 sales 가 포함된 모든 파일명 찾아 오기
all_files = glob.glob(os.path.join(input_path,'*sales*'))

df_list = []

#모든 파일명에서 개별 파일명 하나씩을 꺼냄
for file in all_files:
    #데이터프레임을 만들고(인덱스 컬럼을 만들지 않는다.)
    df= pd.read_csv(file, index_col = None)
    #리스트에 데이터프레임을 추가한다. 
    df_list.append(df)
    
#판다스의 기능을 이용해 리스트에 있는 데이터프레임을 합친다.
# axis =  0 : 위 아래 결합
df_concat = pd.concat(df_list, axis=0 , ignore_index = True)
print(df_concat)
#CSV 파일 작성
df_concat.to_csv(output_file, index = False)
```

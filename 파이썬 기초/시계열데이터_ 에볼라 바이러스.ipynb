```python
#===================================================
#판다스 : 시계열 데이터 다루기
#===================================================

import pandas as pd


input_file = 'E:/Ella/RStudio/DataAnalysis/프로젝트/country_timeseries.csv'

df = pd.read_csv(input_file)
print(df)

#데이터 속성 변경
df.dtypes
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes

#-----------------------------------------------
#다음은 특정 데이터를 인덱스로 접근해야 하기 때문에 불편
df['Date'][3].year
df['Date'][3].month
df['Date'][3].day

#dt 속성값을 이용해 컬럼 추가 - 추가된 컬럼의 자료형은 정수형
df['year']= df['Date'].dt.year
df['month']=df['Date'].dt.month
df['day']=df['Date'].dt.day

df[['Date', 'year', 'month', 'day']]

#에볼라 최초 발병일 계산 : 원 데이터가 시간 역순으로 정렬되어있다.
df.iloc[:, :4].tail()
df.iloc[-5:,:4]

#최초 발병일 찾기
df.iloc[:, :4].tail(1)
df['Date'].min()

#에볼라의 진행 정도 파악 : Day - 기존 데이터
df['outbreak_d'] = df['Date'] - df['Date'].min()
df[['Date', 'Day', 'outbreak_d']].head()
df[['Date', 'Day', 'outbreak_d']].tail()

#누락된 데이터 확인
#2015/01/01, 2014/03/23 ...

#---------------------------------------------------
#나라별로 에볼라의 확산 속도를 비교
#각 나라의 발병일이 달라 그래프가 그려지기 시작한 지점이 다르다.

import matplotlib.pyplot as plt

df.index=df['Date']

fig, ax = plt.subplots()
ax=df.iloc[:, 1:18].plot()
ax.legend(fontsize=7, loc=2, borderaxespad=0.)
plt.show()

#1. 다음 데이터에서 시작한 날이 다름을 확인
#2. 누락된 데이터 확인 : 2014/03/23
df[['Day', 'Cases_Guinea', 'Cases_Liberia']].tail(10)

#---------------------------------------------------
#나라별로 에볼라 확산 속도를 비교하기 위해 발생하기 시작한 날짜 조정

#date 열의 최댓값과 최소값으로 시간 범위 생성
new_idx=pd.date_range(start=df.index.min(), end=df.index.max())
print(new_idx)

#생성된 인덱스 반대로
new_idx = reversed(new_idx)

#새로 생성된 인덱스 적용
df=df.reindex(new_idx)

print(df.head().iloc[:,:4])
print(df.tail().iloc[:,:4])

#---------------------------------------------------
#각 나라의 에볼라 발병일 옮기기
# last_valid_index 는 가장 오래된 데이터를 가져오고
# first_valid_index는 가장 최근의 데이터를 가져온다.=

last_valid = df.apply(pd.Series.last_valid_index)
print(last_valid)
first_valid = df.apply(pd.Series.first_valid_index)
print(first_valid)


#각 나라의 에볼라 발병일을 동일한 출발선으로 옮기기
#에볼라가 가장 처음으로 발병한 날에서 각 나라의 에볼라 발병일을
#뺀 만큼 옮기면 된다.

earliest_date = df.index.min()
print(earliest_date)

shift_values = last_valid - earliest_date
print(shift_values)

ebola_dict= {}
for idx, col in enumerate(df):
    d = shift_values[idx].days
    shifted = df[col].shift(d)
    ebola_dict[col]=shifted
    
ebola_shift = pd.DataFrame(ebola_dict)

#에볼라 최초 발병일을 기준으로 모든 열의 데이터가 옮겨졌다.
ebola_shift.tail()
ebola_shift[['Day', 'Cases_Guinea', 'Cases_Liberia']].tail(10)

#인덱스를 Day로 지정하고, 그래프에 필요없는 Date, Day열 삭제
ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Date', 'Day'], axis=1)

cols=range(16,20)
ebola_shift.columns[cols]
ebola_shift = ebola_shift.drop(ebola_shift.columns[cols], axis=1)
ebola_shift.info()
print(ebola_shift.tail())

#그래프 다시 그리기
fig, ax = plt.subplots()
ax= ebola_shift.iloc[:,:9].plot()
ax.legend(fontsize=7, loc=2, borderaxespad=0)
plt.show()
```
